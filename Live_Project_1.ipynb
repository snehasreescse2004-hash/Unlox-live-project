```python
import pandas as pd
import numpy as np

df = pd.read_csv("bangalore_tech_salaries.csv")

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
role_map = {
    'SDE Backend': 'SDE Backend',
    'Backend Engineer': 'SDE Backend',
    'Backend Dev': 'SDE Backend',
    'BE': 'SDE Backend',
    'SDE BE': 'SDE Backend',

    'SDE Frontend': 'SDE Frontend',
    'Frontend Engineer': 'SDE Frontend',
    'Frontend Dev': 'SDE Frontend',

    'SDE Full Stack': 'SDE Full-Stack',
    'Full Stack Developer': 'SDE Full-Stack',
    'Fullstack Engineer': 'SDE Full-Stack'
}

df["role_clean"] = df["role"].replace(role_map)

print(df["role_clean"].value_counts())
```

```python
df["company_type_clean"] = (
    df["company_type"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "unicorn": "Unicorn",
        "mnc": "MNC",
        "mid-size": "Mid-size",
        "mid size": "Mid-size",
        "early-stage": "Early-stage",
        "early stage": "Early-stage"
    })
    .str.title()
)

print(df["company_type_clean"].value_counts())
```

```python
df["education_tier_clean"] = df["education_tier"].astype(str).str.strip().replace({
    "1": "Tier 1", "T1": "Tier 1", "Tier-1": "Tier 1",
    "2": "Tier 2", "T2": "Tier 2", "Tier-2": "Tier 2",
    "3": "Tier 3", "T3": "Tier 3", "Tier-3": "Tier 3"
})

print(df["education_tier_clean"].value_counts())
```

```python
def clean_ctc(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower().replace("₹", "").replace(" ", "")

    if "lpa" in x:
        return float(x.replace("lpa", ""))

    if "," in x:
        return float(x.replace(",", "")) / 100000

    try:
        val = float(x)
        if val < 1000:
            return val
        else:
            return val / 100000
    except:
        return np.nan
```

```python
df["current_ctc_lpa"] = df["current_ctc"].apply(clean_ctc)
df["previous_ctc_lpa"] = df["previous_ctc"].apply(clean_ctc)

df = df.dropna(subset=["current_ctc_lpa"])

print(df[["current_ctc_lpa", "previous_ctc_lpa"]].head())
```

```python
df = df.drop_duplicates()

df_clean = df.copy()

print(df_clean.shape)
```

```python
# CTC distribution by role

role_ctc = df_clean.groupby("role_clean")["current_ctc_lpa"].agg(["mean", "median", "min", "max"]).sort_values("median", ascending=False)
role_ctc
```

```python
# Experience curve for SDE Backend

backend = df_clean[df_clean["role_clean"] == "SDE Backend"]

backend["exp_band"] = pd.cut(backend["years_exp"], [0,1,3,5,100], labels=["0-1","2-3","4-5","6+"])

backend.groupby("exp_band")["current_ctc_lpa"].median()
```

```python
# Skill premium for SDE

skills = ["AWS", "ML", "System Design", "Kubernetes"]

sde = df_clean[df_clean["role_clean"].str.contains("SDE", na=False)]

for skill in skills:
    with_skill = sde[sde["skills"].str.contains(skill, na=False)]["current_ctc_lpa"].median()
    without_skill = sde[~sde["skills"].str.contains(skill, na=False)]["current_ctc_lpa"].median()

    print(skill, with_skill, without_skill)
```

```python

backend = df_clean[df_clean["role_clean"] == "SDE Backend"]

company_ctc = backend.groupby("company_type_clean")["current_ctc_lpa"].median()

print(company_ctc)

if "Unicorn" in company_ctc.index and "MNC" in company_ctc.index:
    premium = ((company_ctc["Unicorn"] - company_ctc["MNC"]) / company_ctc["MNC"]) * 100
    print("Unicorn vs MNC Premium %:", premium)
else:
    print("MNC or Unicorn not found in dataset")
```

```python

df_clean["exp_band"] = pd.cut(df_clean["years_exp"], [0,1,3,5,100], labels=["0-1","2-3","4-5","6+"])

df_clean["group_median"] = df_clean.groupby(
    ["role_clean", "company_type_clean", "exp_band"]
)["current_ctc_lpa"].transform("median")

df_clean["gap"] = df_clean["current_ctc_lpa"] - df_clean["group_median"]

df_clean.sort_values("gap").head(10)[
    ["role_clean","company_type_clean","years_exp","current_ctc_lpa","group_median","gap"]
]
```
